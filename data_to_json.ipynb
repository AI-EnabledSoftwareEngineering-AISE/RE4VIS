```python
# split t3_ped_df into train, val, and test with 70%, 15%, and 15% respectively

# shuffle the dataframe
t3_ped_df = t3_ped_df.sample(frac=1).reset_index(drop=True)

# split the dataframe into train, val, and test
t3_train_df = t3_ped_df.iloc[:int(0.7*len(t3_ped_df))]
t3_val_df = t3_ped_df.iloc[int(0.7*len(t3_ped_df)):int(0.85*len(t3_ped_df))]
t3_test_df = t3_ped_df.iloc[int(0.85*len(t3_ped_df)):]
print(len(t3_train_df))
print(len(t3_val_df))
print(len(t3_test_df))
```
